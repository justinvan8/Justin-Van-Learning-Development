```python
# Inputting the company name
companyName = input("What is the DSP's Name? ")

# Find the amount of quantity
totalDeliveries = int(input("How many deliveries does the DSP have? "))

# Understand the cost per delivery
costPerDelivery = float(input("Cost Per Delivery Value: "))

# Calculate the total amount
companySales = (totalDeliveries * costPerDelivery)

#print the total amount shown
print("\n--- DSP Delivery Bill Summary ---")
print(f"DSP Name: {companyName}")
print(f"Total Deliveries: {totalDeliveries}")
print(f"Cost Per Delivery: ${costPerDelivery:.2f}")
print(f"Total Delivery Cost: ${companySales:.2f}")
```
